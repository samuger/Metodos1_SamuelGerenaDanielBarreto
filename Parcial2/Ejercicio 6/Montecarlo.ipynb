```python
import numpy as np
import matplotlib.pyplot as plt

def CreateSphere(N,R=1):
    
    Points = np.zeros((N,3))
        
    for i in range(N):
        
        phi = np.random.uniform(0,2*np.pi)
        u = np.random.rand()
        r = R*u**(1/3)
        costheta = np.random.uniform(-1,1)
        theta = np.arccos(costheta)
        
        Points[i] = [r*np.cos(phi)*np.sin(theta),r*np.sin(phi)*np.sin(theta),r*np.cos(theta)]
        
    return Points
```

```python
n=100000
Sphere = CreateSphere(n)
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(Sphere[:,0],Sphere[:,1],Sphere[:,2])
```

```python
x=Sphere[:,0]
y=Sphere[:,1]
z=Sphere[:,2]
```

```python
Ixx=0
suma1=0
for i in range(n):
  suma1+=y[i]**2+z[i]**2
Ixx=(1/n)*suma1
Ixx
```

```python
Iyy=0
suma2=0
for i in range(n):
  suma2+=x[i]**2+z[i]**2
Iyy=(1/n)*suma2
Iyy
```

```python
Izz=0
suma3=0
for i in range(n):
  suma3+=x[i]**2+y[i]**2
Izz=(1/n)*suma3
Izz
```

```python
Ixy=0
suma4=0
for i in range(n):
  suma4+=x[i]*y[i]
Ixy=-(1/n)*suma4
Ixy
```

La esfera, al ser simtrica vista desde todos los planos, nos permite concluir que siempre la inercia ser 2/5 en todos los ejes. 
Los valores de cada inercia varian puesto que el valor de N es finito, si N tendiera a infinito se obtendria 0,4 en la inercia de cada eje.
